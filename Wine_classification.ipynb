```python
data=df
data.head()
```

```python
data.info()
```

```python
X=data.drop('quality', axis='columns')
y=data['quality']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X.head()
```

```python
y.head()
```

```python
X=scaler.fit_transform(X)
```

```python
X
```

```python
X=pd.DataFrame(X, columns=['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=42) 
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(model1.score(X_test, y_test))
```

```python
cm=confusion_matrix(y_test, y_pred)
fig, ax=plt.subplots(figsize=(6,4))
sns.heatmap(cm, annot=True, ax=ax)
```

```python
from sklearn.svm import SVC
model2=SVC()
```

```python
model2.fit(X_train, y_train)
```

```python
y_predx=model2.predict(X_test)
```

```python
confusion_matrix(y_test, y_predx)
```

```python
cm=confusion_matrix(y_test, y_predx)
fig, ax=plt.subplots(figsize=(6,4))
sns.heatmap(cm, annot=True, ax=ax)
```

```python
print(accuracy_score(y_test, y_predx))
```
